```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
from matplotlib import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
def identity_f(s):
    return s
```

```python
def step_f(s):
    return np.sign(s)
```

```python
def sigmoid_f(s):
    return 1/(1+np.exp(-s))
```

```python
def bipolar_sigmoid_f(s):
    return 2/(1+np.exp(-s)) - 1
```

```python
def tanh_f(s):
    return (np.exp(s) - np.exp(-s)) / (np.exp(s) + np.exp(-s))
```

```python
def relu_f(s):
    if s < 0:
        return 0
    else:
        return s
```

```python
ss=np.linspace(-4, 4, 1000)
```

```python
fig, ax = plt.subplots()

#ax.yaxis.set_ticks(np.arange(-1, 1, 0.1))

#ax.plot(ss, identity_f(ss), linewidth=4, label='identity_f')
#ax.plot(ss, [relu_f(s) for s in ss], linewidth=4, label='relu_f')
ax.plot(ss, step_f(ss), linewidth=4, label='step_f')
ax.plot(ss, sigmoid_f(ss), linewidth=4, label='sigmoid_f')
ax.plot(ss, bipolar_sigmoid_f(ss), linewidth=4, label='bipolar_sigmoid_f')
ax.plot(ss, tanh_f(ss), linewidth=4, label='tanh_f')

# set the spine locations
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# turn off the top and right spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.grid(True, which='both')

ax.legend()
```
