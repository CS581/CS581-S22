```python
import matplotlib.pyplot as plt
from matplotlib import pylab
from matplotlib import colors
import numpy as np
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
def set_spines(ax):
    # set the spine locations
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')

    # turn off the top and right spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    ax.grid(True, which='both')
```

# Global

```python
def f(x):
    return -2*(x**2) + 8*x + 10

def fp(x):
    return -4*x + 8
```

```python
x_axis=np.linspace(-3, 7, 100)
_, ax = plt.subplots()
ax.plot(x_axis, f(x_axis), linewidth=3, label='f', color='b')
ax.plot(x_axis, fp(x_axis), linewidth=3, label='fp', color='r')
set_spines(ax)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
ax.legend()
```

```python
lr=0.1
eps = 0.001
max_it = 100

x=1

all_xs = []
all_xs.append(x)

for i in range(max_it):    
    print("Iter %d:\tx:\t%0.10f\tf(x):\t%0.10f\tfp(x):\t%0.10f\tlr*fp(x):\t%0.10f" %(i, x, f(x), fp(x), lr*fp(x)))  
    x += lr*fp(x)
    all_xs.append(x)
    if abs(fp(x)) < eps:
        print("Final:\tx:\t%0.10f\tf(x):\t%0.10f\tfp(x):\t%0.10f\tlr*fp(x):\t%0.10f" %(x, f(x), fp(x), lr*fp(x)))   
        break
```

```python
x_axis=np.linspace(-1, 5, 100)
_, ax = plt.subplots()
ax.plot(x_axis, f(x_axis), linewidth=3, color='b')
ax.plot(all_xs, [f(x_) for x_ in all_xs], 'r+', markersize=20, markeredgewidth=2)
set_spines(ax)
```

# Local

```python
def f(x):
    return -(x**5)-2*(x**4)+13*(x**3)+14*(x**2)-24*x

def fp(x):
    return -5*(x**4)-8*(x**3)+39*(x**2)+28*x-24
```

```python
x_axis=np.linspace(-3.7, 2.7, 100)

fig, ax = plt.subplots()

ax.plot(x_axis, f(x_axis), linewidth=3, label='f', color='b')
ax.plot(x_axis, fp(x_axis), linewidth=3, label='fp', color='r')

set_spines(ax)
ax.legend()
```

```python
lr=0.01
eps = 0.001
max_it = 50

x=0

all_xs = []
all_xs.append(x)

for i in range(max_it):    
    print("Iter %d:\tx:\t%0.10f\tf(x):\t%0.10f\tfp(x):\t%0.10f\tlr*fp(x):\t%0.10f" %(i, x, f(x), fp(x), lr*fp(x)))    
    x += lr*fp(x)
    all_xs.append(x)
    if abs(fp(x)) < eps:
        print("Final:\tx:\t%0.10f\tf(x):\t%0.10f\tfp(x):\t%0.10f\tlr*fp(x):\t%0.10f" %(x, f(x), fp(x), lr*fp(x)))  
        break
```

```python
x_axis=np.linspace(-3.7, 2.7, 100)
_, ax = plt.subplots()
ax.plot(x_axis, f(x_axis), linewidth=3, color='b')
ax.plot(all_xs, [f(x_) for x_ in all_xs], 'r+', markersize=20, markeredgewidth=2)
set_spines(ax)
```
